```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
z = y - x
```

### DEFINIR ESTILO

```python
plt.style.use('seaborn')  
```

### GERAL

```python
plt.style.use('seaborn')  

plt.xlabel('Título principal')
plt.ylabel('Título do eixo Y')
plt.title('Título principal')

plt.plot(x, x, color = '#E10600',ls= ':', label='x', marker='o', alpha = 1)
plt.plot(x, y, color = 'g',ls= '--',label='x²', marker='*', alpha = 0.5) 
plt.plot(x, z, color ='b',ls= '-',label='x²-x', marker='.', alpha = 0.7) 

plt.legend(loc = 0, frameon = True, borderaxespad=1)

#plt.show()
```

### SUBPLOTS

Automatic axis manager

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_xlabel('aX')
axes[0].set_ylabel('aY')
axes[0].set_title('Gráfico A')
axes[0].plot(x, y, color = '#E10600',ls= ':', label='x', marker='o', alpha = 1)

axes[1].set_xlabel('bX')
axes[1].set_ylabel('bY')
axes[1].set_title('Gráfico B')
axes[1].plot(y, x, color = 'g',ls= '--',label='x²', marker='*', alpha = 0.5) 
```

Definir dimensões da figura

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Titulo');
```
